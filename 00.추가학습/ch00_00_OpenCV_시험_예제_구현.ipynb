```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
from google.colab.patches import cv2_imshow
cv2.__version__
```

1번. 아래와 같은 원을 그리도록 코드를 작성해 보았습니다. x, y, r을 구하세요.

```python
shape = (100,100,3)
value = 255
background = np.full((shape), value, np.uint8)
ans = cv2.circle(background, (50, 50), 30, (0,255,0), 2)
plt.imshow(ans)
# x = 50, y = 50, r = 30
```

2번. 아래 이미지에서 개 이미지만 크롭하기 위해 아래와 같이 코드를 작성해 보았습니다. a,b,c,d에 
들어갈 숫자를 구하세요. 

```python
path = 'horse.jpg'
img = cv2.imread(path)
cv2_imshow(img[300:600, 600:800])
# a = 300, b = 600, c = 600, d = 800
```

3번. 이미지 임계 처리 방법중 아래와 같은 결과를 내는 옵션은 무엇인가요? (titles 중에 고르세요)

```python
line = np.linspace(0,255,100).reshape(100,1)
img2 = np.concatenate([line for i in range(100)],axis=1)
cv2_imshow(img2)

ret, dst = cv2.threshold(img2, 50, 255, cv2.THRESH_BINARY)
print('THRESH_BINARY')
cv2_imshow(dst)
```

7번. circle과 mask가 아래와 같이 이미지로 바인딩 되어있을 때 `bitwise_and` 함수의 결과인 
masked 이미지를 그리세요. 

```python
shape = (100,100,3)
value = 255

bkg1 = np.full((shape), value, np.uint8)
print('circle')
circle = cv2.circle(bkg1, (50,50), 50, (0,255,0), -1)
cv2_imshow(circle)
print(circle.shape)


bkg2 = np.full((100,100), value, np.uint8)
print('mask')
mask = cv2.rectangle(bkg2, (0,50), (50,100), (0,0,0), -1)
mask = cv2.rectangle(bkg2, (50,0), (100,50), (0,0,0), -1)
mask = cv2.rectangle(bkg2, (50,50), (100,100), (0,0,0), -1)
cv2_imshow(mask)
print(mask.shape)

print('masked')
masked = cv2.bitwise_and(circle, circle, mask=mask)
cv2_imshow(masked)
```
